```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Placement_Data_Full_Class.xls")
```

```python
print("--- FIRST 5 ROWS ---")
print(df.head())
```

```python
print("\n--- MISSING VALUE COUNT PER COLUMN ---")
print(df.isnull().sum())
```

```python
df['salary'] = df['salary'].fillna(0)
```

```python
initial_rows = df.shape[0]
df = df.drop_duplicates()
print(f"\nRemoved {initial_rows - df.shape[0]} duplicate rows.")
```

```python
text_cols = ['gender', 'ssc_b', 'hsc_b', 'hsc_s', 'degree_t', 'workex', 'specialisation', 'status']
for col in text_cols:
    df[col] = df[col].astype(str).str.strip().str.title()
```

```python
df['salary'] = df['salary'].astype(int)
```

```python
df = df.rename(columns={
    'sl_no': 'serial_no',
    'ssc_p': 'ssc_percentage',
    'ssc_b': 'ssc_board',
    'hsc_p': 'hsc_percentage',
    'hsc_b': 'hsc_board',
    'hsc_s': 'hsc_stream',
    'degree_p': 'degree_percentage',
    'degree_t': 'degree_type',
    'workex': 'work_experience',
    'etest_p': 'employability_test_percentage',
    'mba_p': 'mba_percentage',
    'status': 'placement_status'
})
```

```python
df['overall_academic_avg'] = round((df['ssc_percentage'] + df['hsc_percentage'] + df['degree_percentage']) / 3, 2)
```

```python
df['salary_bracket'] = np.where(df['salary'] == 0, 'No Package',
                       np.where(df['salary'] < 250000, 'Standard Package', 'Premium Package'))
```

```python
df.to_csv("cleaned_dataset.csv", index=False)
print("\nSuccess: 'cleaned_dataset.csv' exported successfully!")
```
